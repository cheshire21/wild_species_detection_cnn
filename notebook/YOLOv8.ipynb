# 2. Train
First add the custom data file and dataset on Colab file and then train the yolov8x model

```
import time
inicio = time.time()
# Train YOLOv8n on COCO128 for 3 epochs
!yolo train model=yolov8x.pt data=/content/custom_data.yaml epochs=70 imgsz=640 batch=16 project=/content/results name=yolov8x
fin = time.time()
#Tiempo en minutos
tiempo_ejecucion_yolov8x=(fin-inicio)/60
print("El tiempo de ejecuciÃ³n es de: ",tiempo_ejecucion_yolov8x)
```
